```python
import matplotlib.pyplot as plt
```

##СОЗДАЕМ ПУСТЫЕ СПИСКИ

```python
history1 = []
history2 = []
history3 = []
history4 = []
sales_1_history = []
sales_2_history = []
sales_3_history = []
sales_4_history = []
```

##ЗАДАЕМ НАЧАЛЬНЫЕ ЗНАЧЕНИЯ


```python
VOLUME = 500
excerpt = 0.3
sales_1 = 50
sales_2 = 70
sales_3 = 50
sales_4 = 20
prob_1 = 0.9
prob_2 = 0.6
prob_3 = 0.5
prob_4 = 0.3
sale_rate_1 = prob_1*sales_1*(1-(sales_1+sales_2+sales_3+sales_4)/VOLUME)
sale_rate_2 = prob_2*sales_2*(1-(sales_1+sales_2+sales_3+sales_4)/VOLUME)
sale_rate_3 = prob_3*sales_3*(1-(sales_1+sales_2+sales_3+sales_4)/VOLUME)
sale_rate_4 = prob_4*sales_4*(1-(sales_1+sales_2+sales_3+sales_4)/VOLUME)
sector_1 = sales_1 / (sales_1 + sales_2 + sales_3 + sales_4)
sector_2 = sales_2 / (sales_1 + sales_2 + sales_3 + sales_4)
sector_3 = sales_3 / (sales_1 + sales_2 + sales_3 + sales_4)
sector_4 = sales_4 / (sales_1 + sales_2 + sales_3 + sales_4)
price_change_1 = -excerpt*(1 - sector_1)
price_change_2 = -excerpt*(1 - sector_2)
price_change_3 = -excerpt*(1 - sector_3)
price_change_4 = -excerpt*(1 - sector_4)
history1.append(sector_1)
history2.append(sector_2)
history3.append(sector_3)
history4.append(sector_4)
sales_1_history.append(sales_1)
sales_2_history.append(sales_2)
sales_3_history.append(sales_3)
sales_4_history.append(sales_4)
for t in range(1, 20):
    sales_1 += sale_rate_1
    sales_2 += sale_rate_2
    sales_3 += sale_rate_3
    sales_4 += sale_rate_4
    sector_1 = sales_1 / (sales_1 + sales_2 + sales_3 + sales_4)
    sector_2 = sales_2 / (sales_1 + sales_2 + sales_3 + sales_4)
    sector_3 = sales_3 / (sales_1 + sales_2 + sales_3 + sales_4)
    sector_4 = sales_4 / (sales_1 + sales_2 + sales_3 + sales_4)
    price_change_1 = -excerpt*(1 - sector_1)
    price_change_2 = -excerpt*(1 - sector_2)
    price_change_3 = -excerpt*(1 - sector_3)
    price_change_4 = -excerpt*(1 - sector_4)
    prob_1 = 0.9 + price_change_1
    prob_2 = 0.6 + price_change_2
    prob_3 = 0.5 + price_change_3
    prob_4 = 0.3 + price_change_4
    sale_rate_1 = prob_1*sales_1*(1-(sales_1+sales_2+sales_3+ sales_4)/VOLUME)
    sale_rate_2 = prob_2*sales_2*(1-(sales_1+sales_2+sales_3+ sales_4)/VOLUME)
    sale_rate_3 = prob_3*sales_3*(1-(sales_1+sales_2+sales_3+ sales_4)/VOLUME)
    sale_rate_4 = prob_4*sales_4*(1-(sales_1+sales_2+sales_3+ sales_4)/VOLUME)
    history1.append(sector_1)
    history2.append(sector_2)
    history3.append(sector_3)
    history4.append(sector_4)
    sales_1_history.append(sales_1)
    sales_2_history.append(sales_2)
    sales_3_history.append(sales_3)
    sales_4_history.append(sales_4)
```

```python
from matplotlib.pyplot import figure
```

```python
figure(figsize=(18, 6), dpi=80)
plt.plot(range(20), history1)
plt.plot(range(20), history2)
plt.plot(range(20), history3)
plt.plot(range(20), history4)
plt.grid()
plt.legend(['sector_1', 'sector_2', 'sector_3', 'sector_4'])
```

```python
figure(figsize=(18, 6), dpi=80)
plt.plot(range(20), sales_1_history)
plt.plot(range(20), sales_2_history)
plt.plot(range(20), sales_3_history)
plt.plot(range(20), sales_4_history)
plt.grid()
plt.legend(['sector_1', 'sector_2', 'sector_3', 'sector_4'])
```
