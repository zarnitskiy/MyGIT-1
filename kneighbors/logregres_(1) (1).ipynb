```python
DATA = pd.read_csv("report.csv")
DATA
```

```python
DATA["MIP"].mean()
```

```python
p = ['MIP',	'STDIP', 'EKIP', 'SIP', 'MC', 'STDC', 'EKC', 'SC']
r = ['MIPnorm',	'STDIPnorm', 'EKIPnorm', 'SIPnorm', 'MCnorm', 'STDCnorm', 'EKCnorm', 'SCnorm']
for j in range(0, len(p)):
    z = []
    xmax = DATA[p[j]].max()
    xmin = DATA[p[j]].min()
    for i in range(0,len(DATA[p[j]])):
        x = DATA[p[j]][i]
        z.append((x-xmin)/(xmax-xmin))
    DATA[r[j]]=z
print(DATA.MIPnorm.mean())
DATA.head(7)
```

```python
X = pd.DataFrame(DATA[r])
```

```python
y = pd.DataFrame(DATA['TARGET'])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression(random_state=2019, solver='lbfgs').fit(X, y.values.ravel())
```

```python
reg.predict_proba([[0.539, 0.162, 0.188, 0.653, 0.796, 0.233, 0.397, 0.065]])
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5, p=2)
neigh.fit(X, y)
```

```python
NewObject = [0.539, 0.162, 0.188, 0.653, 0.796, 0.233, 0.397, 0.065]
neigh.predict([NewObject])
```
