(а) обыкновенного дифференциального уравнения; 



```python
N_0 = [15] 
birth_rate  = 1.3 
death_rate = 6 
delta_time = 0.01
total_time = int(4/delta_time)
for t in range(0, total_time ):
    updated_N_0 = N_0[t] + delta_time * (birth_rate * N_0[t]  - death_rate  * N_0[t])
    N_0.append(updated_N_0)

time_points = range(total_time + 1)
plt.plot(time_points, N_0) 
print(N_0[400])
```
